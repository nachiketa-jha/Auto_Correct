```python
df = pd.read_csv('cricket_data_2025.csv')
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df_cleaned = df.dropna(subset=['Year'])
```

```python
import numpy as np
features =['Matches_Batted','Not_Outs','Highest_Score','Batting_Average','Balls_Faced','Batting_Strike_Rate','Centuries','Half_Centuries','Fours','Sixes']
df.replace('No stats', np.nan, inplace=True)
df[features] = df[features].apply(pd.to_numeric, errors='coerce')

df_cleaned = df.dropna(subset=features + ['Runs_Scored']).reset_index(drop=True)
```

```python
df_cleaned.head(10)
```

```python
from sklearn.preprocessing import MinMaxScaler
# df_cleaned['Highest_Score']=df_cleaned['Highest_Score'].str.replace('*','')
scaler = MinMaxScaler()

X=df_cleaned[features]
y=df_cleaned['Runs_Scored']
X_scaled = scaler.fit_transform(X)
```

```python
assert len(X_scaled) == len(y)
df_cleaned['Runs_Scored'].value_counts().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
predictions = model.predict(X_test)

mae = mean_absolute_error(y_test, predictions)
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print(f"Model Evaluation:\nMAE: {mae}\nMSE: {mse}\nR2 Score: {r2}")
```

```python
df_cleaned['Predicted_Runs'] = model.predict(scaler.transform(df_cleaned[features]))
top_scorer = df_cleaned.loc[df_cleaned['Predicted_Runs'].idxmax(), 'Player_Name']

print(f"\n🏆 Predicted Highest Run Scorer: {top_scorer}")
```

```python
new_features = ['Matches_Bowled','Balls_Bowled','Runs_Conceded','Bowling_Average','Economy_Rate','Bowling_Strike_Rate','Four_Wicket_Hauls','Five_Wicket_Hauls']
df.replace('No stats', np.nan, inplace=True)
```
